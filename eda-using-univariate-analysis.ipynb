## Univariate Analysis 
To analysis in one column called univariate Analysis
Based on two types of Data:
 1. Numerical Data (For ex:- Height , weight, age, phone price)
 2. Categorical Data
    For ex:- (Nationality,Gender, Branch)

## 1.Categorical Data:-
a) Countplot

```python
sns.countplot(df['Survived'])  
```

## 0R WE CAN USE 

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
df['Embarked'].value_counts().plot(kind='bar')
```

## B) PIECHART 
IF WE WANT DATA IN TERMS OF PERCENTAGE THEN WE USE PIECHART
 ## AUTOPCT IS 
 used for to show the percentages in pie chart

```python
df['Embarked'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Age'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

# WE FIND ANY CATEGORICAL DATA WITH THEIR IT IS HAVING ONLY THESE TWO OPTIONS IN CATEGORICAL DATA of univariate analysis

## a) HISTOGRAM

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'])
```

```python
 ## WE USE BINS TO ADJUST OUR ACCUURACY 
plt.hist(df['Age'], bins = 30)
```

## B) Distplot:-


 Distplot is a visualization technique used to show the distribution of a dataset by combining a histogram with a kernel density estimation (KDE) curve, helping to understand how data values are spread and where they are concentrated. 

it also define the skewness

```python
sns.distplot(df['Age'])
```

```python
sns.distplot(df['Survived'])
```

## C) BoxPlot : 
Boxplot is a statistical graph used to represent the distribution of numerical data by showing its minimum, first quartile (Q1), median (Q2), third quartile (Q3), maximum, and outliers.


It is used on Numerical Data

```python
sns.boxplot(df['Fare'])
```

```python
sns.boxplot(df['Pclass'])
```

```python
sns.boxplot(df['Embarked'])
```
