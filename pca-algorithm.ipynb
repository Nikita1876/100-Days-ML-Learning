### PCA ALGORITHM

it is a mathematical technique to reduce column but can take summarize important information of that data.

```python
from sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Age': [10,20,45,56,67,33],
    'Income':[30000,40000,56000,78000,90000,76000],
    'Spending':[700,4000,6000,5500,3000,1000]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# New DataFrame
pca_df = pd.DataFrame(pca_result, columns=['PCA1','PCA2'])

# Variance
explained_variance = pca.explained_variance_ratio_
print(np.round(explained_variance * 100, 2))

# Plot
plt.figure(figsize=(8,6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], color='black', s=80)

plt.title("PCA Projection (2D View)")
plt.xlabel('PCA1 (Main Pattern)')
plt.ylabel('PCA2 (Minor Pattern)')
plt.grid(True)
plt.show()
print ("NEW DATA WITH 2 FEATURES ")
```
