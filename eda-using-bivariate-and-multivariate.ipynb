```python
## LET'S SAY WE HAVE TO FIND AVERAGE AGE ON THE BASIS PF AGE CLASS 
## Relationship between Pclass and Age 
sns.barplot(x=titanic['Pclass'],y=titanic['Age'])
```

## Using MultiVariate Analysis

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

## BOXPLOT(NUMERICAL-CATEGORICAL)

```python
sns.boxplot(x=titanic['Sex'], y=titanic['Age'])
```

```python
sns.boxplot(x='Sex', y='Age', data=titanic, palette=['blue', 'pink'])
```

## BY USING MULTIVARIATE ANALYSIS

```python
sns.boxplot(x='Sex', y='Age', data=titanic, palette=['blue', 'pink'],hue=titanic['Survived'])
```

## DistPlot     (NUMERICAL-CATEGORICAL)

```python
sns.distplot(titanic['Age'])
```

```python
## USING MULTIVARIATE ANALYSIS WITHOUT USING THE HUE IN DISTPLOT
sns.distplot(titanic[titanic['Survived']== 0]['Age'],hist = False)
sns.distplot(titanic[titanic['Survived']== 1]['Age'],hist = False)
```
