```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/kaggle/input/datasets/brendan45774/test-file/tested.csv')
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace = True)
```

```python
df.head()
```

```python
## St ep -> 1 TRAIN_TEST_SPLIT
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                df['Survived'],
                                                test_size=0.2,
                                                random_state=42)
```

```python
X_train.head(2)
```

```python
y_train.head()
```

```python
df.isnull().sum()
```

```python
## Applying Imputation

si_age = SimpleImputer()
si_fare = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_fare = si_fare.fit_transform(X_train[['Fare']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_fare= si_fare.fit_transform(X_test[['Fare']])
```

```python
X_train.head(2)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe_sex = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Train
X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train[['Embarked']])

# Test
X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test[['Embarked']])
```

```python
X_train_embarked
```

```python
X_train_sex
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_sex,X_train_embarked),axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_sex,X_test_embarked),axis=1)
```

```python
X_train_transformed
```

```python
X_test_transformed.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
