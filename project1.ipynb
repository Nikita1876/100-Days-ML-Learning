```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

df = pd.read_csv('/kaggle/input/datasets/aryashah2k/sample-students-grades-dataset-tutorial-notebook/grades.csv')
df
```

```python
df.isnull().mean()
```

```python
df['grade'] = df['grade'].fillna(df['grade'].mean())
```

```python
df
```

```python
X = df[['exam']]   # input feature
y = df[['grade']]  # correct target column

model = LinearRegression()
model.fit(X, y)

predicted_score = model.predict(X)
mae = mean_absolute_error(y,predicted_score)
mse = mean_squared_error(y,predicted_score)
rmse = np.sqrt(mse)

print(mae)
print(mse)
print(rmse)


new_prediction = model.predict([[7]])
print("Predicted Score {new_pred} hour ", new_prediction)
```
